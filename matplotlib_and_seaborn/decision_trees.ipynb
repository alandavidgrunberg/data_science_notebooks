```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
df['Male'] = df['Sex'] == 'male'

## plotting decision tree with two features

features = ['Pclass', 'Male'] # features as separate list of strings
X = df[features].values
y = df['Survived'].values

dt = DecisionTreeClassifier()
dt.fit(X, y) # fitting tree with whole dataset, no train/test split or cross folding

fig = plt.figure(figsize=(15,10)) # customize figure size to ensure tree is big enough to read
tree.plot_tree(dt, feature_names=features) # plot tree, passing 'dt' the decision tree model, 'features' list of strings to 'feature_names="
plt.show()

# reading the plot: we see for example, that 94 passengers were first-class females. Of the first-class females, 3 died and 91 lived
```

```python
## plotting decision tree with two features, different plotting settings 

features = ['Pclass', 'Male'] 
X = df[features].values
y = df['Survived'].values

dt = DecisionTreeClassifier()
dt.fit(X, y)

fig = plt.figure(figsize=(15,10)) 
tree.plot_tree(dt, feature_names=features, class_names=['die','live'], filled=True, proportion=True) 
# passing list of strings ['die','live'] to 'class_name=' this labels predicted outcomes at node (final predictions at leaf nodes)
# passing True to 'filled=' colors show purity of nodes
# passing True to 'proportion=' to display samples and values as proportions instead of # of datapoints
    # for example, we see that 10.6% of the dataset first-class females. Of the first-class females, 3.2% died and 96.8% lived
plt.show()

# code and comments by github.com/alandavidgrunberg
```
