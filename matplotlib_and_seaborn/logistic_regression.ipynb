```python
import matplotlib.pyplot as plt
import pandas as pd # pandas stands for Panel Data
import numpy as np # numpy stands for Numerical Python

df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv') # 'read_csv()' takes a file in csv (comma seperated value) format and converts it to a Pandas DataFrame

plt.scatter(df['Age'], df['Fare'], c=df['Pclass']) # scatter plot, x-axis 'Age', y-axis 'Fare', 'c=' colors the datapoints according to 'Pclass' (1,2,3)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.plot([0, 80], [85, 5]) # plot a line from (0, 85) to (80, 5) that roughly seperates first class from second & third class (manually, not algorithmically)
                           # notice the syntax is [x1, x2], [y1, y2] for points (x1, y1) and (x2, y2)
plt.show()
```

```python
plt.scatter(df['Fare'], df['Age'], c=df['Survived']) # 'Fare' vs 'Age' colored according to 'Survived'
plt.xlabel('Fare')
plt.ylabel('Age')
plt.plot([30,110],[0,80]) # plot a line from (30, 0) to (110, 80) that roughly seperates non-survivors on left from survivors on right                      
plt.show()

# line equation:  0 = ax + by + c
#                 0 = (1)x + (-1)y + (-30)
# any point on the line will fit the equation. for example (50, 20)
#                 1(50) + -1(20) + (-30) = 0
# any point to the left will give a negative number. for example (10, 50)
#                 1(10) + -1(50) + (-30) = -70
# any point to the right will give a positive number. for example (100, 20)
#                 1(100) + -1(20) + (-30) = 50
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression # import model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')

## building logistic regression model (all 6 features)

# prepping data
df['Male'] = df['Sex'] == 'male' # create a new column 'Male' with boolean True/False values to tell us if passenger is male or not 
                                 # booleans are easier for Python to do computations on than the strings in the 'Sex' column

# feature selection
X = df[['Pclass', 'Male', 'Age', 'Siblings/Spouses', 'Parents/Children', 'Fare']].values
y = df['Survived'].values

# instantiate model
model = LogisticRegression()

# fit
model.fit(X, y)

# predict
y_pred = model.predict(X) # array of predictions

# evaluate model
print(confusion_matrix(y, y_pred)) # output confusion matrix, passing actual and predicted targets
plot_confusion_matrix(model, X, y, values_format='d', cmap='Greens') # plot confusion matrix, pass model, input values X, actual target values
                                                      # values_format='d' to correctly format numbers the matrix shows (otherwise uses scientific notation)
                                                      # cmap='Greens' to choose colormap
plt.show()

# scikit-learn convention for confusion matrix is to show predicted labels along columns, actual labels along rows
# scikit-learn convention considers 0 to be negative class and 1 to be positive class

# order of 0 and 1 in confusion matrix depends on which appears first in your dataset/target values
# in our dataset/target values 0 appears first, so 0 is first row and column in above confusion matrix

# Therefore, above confusion matrix shows:

#     A 0     TN | FP
#     c       ___|___
#     t          |
#     u 1     FN | TP  
#     a      
#     l       0     1
#          P r e d i c t
```

```python
# order of 0 and 1 in confusion matrix can be overrided with 'labels='
# lets make 1 appear first with 'labels=[1,0]'
print(confusion_matrix(y, y_pred, labels = [1,0]))
plot_confusion_matrix(model, X, y, labels = [1,0], values_format='d', cmap='Greens') 
plt.show()

# Now, confusion matrix shows:

#     A 1     TP | FN
#     c       ___|___
#     t          |
#     u 0     FP | TN  
#     a      
#     l       1     0 
#          P r e d i c t
```

```python
# To flip matrix so predicted labels along rows, actual labels along columns, we must pass predicted labels first, actual labels second
print(confusion_matrix(y_pred, y, labels = [1,0])) # 'y_pred' passed first, 'y' passed second 
# **Can't figure out how to flip plot_confusion_matrix, not sure if possible**

# Now, confusion matrix shows:

#     P 
#     r 1     TP | FP
#     e       ___|___         
#     d          |
#     i 0     FN | TN     
#     c             
#     t       1     0
#           A c t u a l
```

```python
# customized confusion matrix with TP, FP, TN, FN labels using Seaborn
# **not my code, figure out how this works!**
import seaborn as sns
mat = confusion_matrix(y, y_pred) 
mat_labels = ['TN','FP','FN','TP']
mat_counts = ["{0:0.0f}".format(value) for value in mat.flatten()]
labels = [f"{ii}\n\n{jj}" for ii, jj in zip(mat_labels,mat_counts)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(mat, square=True, annot=labels, cmap='Blues', fmt='')
plt.title('Confusion Matrix', weight='bold', fontsize=16)
plt.ylabel('Actual label', weight='bold')
plt.xlabel('Predicted label', weight='bold')
plt.show()

# code and comments by github.com/alandavidgrunberg
```
