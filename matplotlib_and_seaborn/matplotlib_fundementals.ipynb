```python
import numpy as np
import matplotlib.pyplot as plt
import math

#%matplotlib inline #resets graph style to inline default

x = np.linspace(0,10,1000) #1darray of 1000 values between 0 and 10, evenly spaced

#The figure size (if not set, defaults smaller)
plt.figure(figsize=(10, 6)) #setting size in inches

#plotting the functions
plt.plot(x, np.cos(x), label='cosin') #label will show up in legend
plt.plot(x, np.sin(x), label='sin') #label will show up in legend

#The title of graphic
plt.title('figure1')
#axes in the graphic
plt.xlabel('axe x')
plt.ylabel('axe y')

plt.legend(loc = "lower left") #add the name of your legend in function1 and function2
plt.show()
```

```python
x = np.linspace(0,10,1000) 
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function sin(x)')

# plt.xlim() and plt.ylim() "zoom in" on specified area of graph
plt.xlim(2,7)
plt.ylim(-.5, .75)
plt.show()
```

```python
x = np.linspace(0,10,1000)

#plottng with custom colors and line styles
plt.plot(x, np.cos(x), color = "tab:purple")
plt.plot(x, x/5, color = 'seagreen', linestyle = ":")
plt.plot(x, np.sin(x), "k--", alpha= 0.3)
plt.show()
```

```python
x = np.linspace(0,10,1000) # x is 1000 evenly spaced numbers from 0 to 10
y = np.sin(x)

a = np.linspace(3,7,1000) # a is 1000 evenly spaced numbers from 3 to 7
b = a

import matplotlib.pyplot as plt

#creating graph using plt.plot()
plt.plot(x,y)
plt.plot(a,b)
plt.show()
```

```python
#creating graph using plt.figure() & plt.axes()
#using .twinx() to make a different y axis for each line 
fig2 = plt.figure()
ax2 = plt.axes()
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2twin = ax2.twinx()
ax2twin.set_ylabel('y2 label')
ax2.plot(x,y)
ax2twin.plot(a,b)
plt.show()
```

```python
#creating graph using plt.subplots()
#using .twiny() to make a different x axis for each line
fig3, ax3 = plt.subplots()
ax3.set_xlabel("x label")
ax3.set_ylabel("y label")
ax3twin = ax3.twiny()
ax3twin.set_xlabel ("x2 label")
ax3.plot(x,y)
ax3twin.plot(a,b)
plt.show()
```

```python
#different x & y axis for each line
fig4, ax4 = plt.subplots()
ax4twin = ax4.twinx().twiny()
ax4.plot(x,y)
ax4twin.plot(a,b)
plt.show()
```

```python
x = np.linspace(0,10,1000) 
y = np.sin(x)

a = np.linspace(3,7,1000)
b = a

#different x & y axis for each line
#labeling axes when there are two different x axes and y axes
fig, ax = plt.subplots()
ax.set_xlabel('ax x label')
ax.set_ylabel('ax y label')
axtwo = ax.twinx().twiny()
axtwo.set_xlabel('axtwo x label')
axtwo.set_ylabel('axtwo y label') #not showing up, why?
ax.plot(x,y)
axtwo.plot(a,b)
plt.show()
```

```python
import random as rd
import matplotlib.pyplot as plt

#making x a list of 15 random integers between 0 and 10
x = []
for i in range(15):
  x.append(rd.randint(0,10))


#making y a list of 15 random integers between 66 and 99 
y = []
for i in range(15):
  y.append(rd.randint(66,99))

#making a scatter plot of x & y using plt.scatter()
plt.scatter(x,y)
plt.show()
```

```python
#making a scatter plot of x & y passing 'o' in plt.plot() does the same thing
plt.plot(x,y,'o')
plt.show()
```

```python
#customing the scatter plot; changing the markers and adding labels
plt.scatter(x, y, marker = '*', color='r')
plt.xlabel('age'); 
plt.ylabel('weight')
plt.title('Cats')
plt.show()
```

```python
#same data as a line graph instead of a scatter plot, looks crazy!
plt.plot(x,y)
plt.show()
```

```python
import pandas as pd

#building data frame

#first building dictionary with column titles as keys and arrays as arguments
alcohol_dict = {
  'red': [3, 6, 5, 12, 18, 3, 7, 9, 17, 5], 
  'white': [5, 6, 10, 23, 11, 25, 1, 2, 4, 16],
  'pink': [1, 2, 18, 5, 21, 5, 3, 13, 19, 20], 
  'favorite beer': ['chang','chang','tiger','singha','leo','chang','singha','leo','chang','leo']}

#then building DataFrame by passing the dictionary and an index with a list of row titles
drinks_df = pd.DataFrame(alcohol_dict, index=["adam", "bob", "charles", "james", "maria", "leslie", "alan", "pablo", "shufu", "kevin"])

print(drinks_df)
```

```python
#plotting scatter plot directly from the DataFrame with .plot & kind='scatter'
drinks_df.plot(x = 'red', y = 'white', kind='scatter', marker='x', title="drinking habits", figsize=(5,5))
plt.show()
```

```python
#plotting histogram directly from the DataFrame
#with .plot & kind='hist'
drinks_df['pink'].plot(kind='hist', title = 'pink drinks', bins=5)
plt.show()
```

```python
#with .hist
drinks_df['pink'].hist(bins=5)
plt.show()
```

```python
#plotting histogram with plt.hist
#plt.hist() also returns array of frequency and bin endpoints
x = plt.hist(drinks_df['red'], bins=4)
plt.title('red drinks')
plt.ylabel('how frequent')
plt.show()
print()
print(x)
```

```python
#plotting boxplot directly from the DataFrame with .boxplot
#box is interquartile range (IQR) between the first quartile and the third
#50% of data fall in this range^
#median line inside the box.
#minimum and maximum lines at bottom and top.
drinks_df.boxplot(column = 'white')
plt.show()
```

```python
#plotting boxplot with plt.boxplot
plt.boxplot(drinks_df['white'])
plt.show()
```

```python
#plotting bar plot of favorite beer directly from the DataFrame with .plot & kind='bar'
beer_count = drinks_df['favorite beer'].value_counts() #returns a series (labeled 1d array) of number of people who faved each beer
beer_count.plot(kind="bar")
plt.show()

# code and comments by github.com/alandavidgrunberg
```
